```python
from numpy.random import randn
import seaborn as sns
import numpy as np
from scipy import stats
%matplotlib inline
```

```python
data1=randn(100)
data2=randn(100)+2
```

```python
sns.distplot(data1)
sns.distplot(data2)
```

```python
sns.boxplot(data=[data1,data2])
```

```python
sns.boxplot(data=[data1,data2],whis=np.inf)
#外れ値をひげに含む
```

```python
sns.boxplot(data=[data1,data2],orient="h",whis=np.inf)
```

```python
#ヴァイオリンプロット
data1=stats.norm(0,5).rvs(100)
#正規分布に従う　乱数　　平均が0 標準偏差が5の数字を100個
```

```python
data2=np.concatenate([stats.gamma(5).rvs(50)-1,-1*stats.gamma(5).rvs(50)])
```

```python
sns.boxplot(data=[data1,data2])
```

```python
sns.violinplot(data=[data1,data2])
#カーネル密度推定を使って作られている bwを変えられる
```

```python
sns.violinplot(data=data2,bw=0.01)
```

```python
sns.violinplot(data=data1,inner="stick")
#データの位置が線で示される
```
