```python
import pandas as pd
from pandas import Series,DataFrame
```

```python
titanic_df=pd.read_csv("train.csv")
#sibSp 兄弟と一緒　Parch 親もしくは兄弟と一緒 #Embarked どこから乗船したか
```

```python
#どういう人たちが乗船していたか
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
sns.countplot("Sex",data=titanic_df)
#男性が圧倒敵に多い
```

```python
sns.countplot("Sex",data=titanic_df, hue="Pclass")
```

```python
sns.countplot("Pclass",data=titanic_df, hue="Sex")
```

```python
def male_female_child(passenger):
    age,sex=passenger
    if age<16:
        return "child"
    else:
        return sex

titanic_df["person"]=titanic_df[["Age","Sex"]].apply(male_female_child,axis=1)
```

```python
titanic_df.head(10)
```

```python
sns.countplot("Pclass",data=titanic_df,hue="person")
```

```python
titanic_df["Age"].mean()
```

```python
titanic_df["person"].value_counts()
```

```python
fig = sns.FacetGrid(titanic_df,hue="Sex",aspect=4) #aspect グラフの長さ
fig.map(sns.kdeplot,"Age",shade=True)
oldest=titanic_df["Age"].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```

```python
fig = sns.FacetGrid(titanic_df,hue="person",aspect=4) #aspect グラフの長さ
fig.map(sns.kdeplot,"Age",shade=True)
oldest=titanic_df["Age"].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```

```python
fig = sns.FacetGrid(titanic_df,hue="Pclass",aspect=4) #aspect グラフの長さ
fig.map(sns.kdeplot,"Age",shade=True)
oldest=titanic_df["Age"].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```

```python
#客室のレベルが下がるにつれて年齢層は下がる
```

```python
titanic_df.head()
```

```python
deck=titanic_df["Cabin"].dropna()
```

```python
levels=[]
for level in deck:
    levels.append(level[0])
```

```python
cabin_df=DataFrame(levels)
cabin_df.columns=["Cabin"]
```

```python
sns.countplot("Cabin",data=cabin_df,palette="winter_d",order=sorted(set(levels)))
```

```python
#Tの外れ値を取り除く
cabin_df=cabin_df[cabin_df.Cabin !="T"]
```

```python
sns.countplot("Cabin",data=cabin_df,palette="summer",order=sorted(set(cabin_df.Cabin)))
```

```python
sns.countplot("Embarked",data=titanic_df,hue="person")
```

```python
sns.countplot("Embarked",data=titanic_df,hue="Pclass")
```

```python
titanic_df["Alone"] = titanic_df.Parch+titanic_df.SibSp
titanic_df["Alone"].head()
```

```python
titanic_df["Alone"].loc[titanic_df["Alone"] > 0] = "With Family"
titanic_df["Alone"].loc[titanic_df["Alone"] == 0] = "Alone"
titanic_df.head(20)
```

```python
sns.countplot("Alone",data=titanic_df,palette="Blues")
```

```python
#どういう人が生還したか
titanic_df["Survivor"]=titanic_df.Survived.map({0:"no",1:"yes"})
```

```python
sns.countplot("Survivor",data=titanic_df,palette="Set1")
```

```python
sns.factorplot("Pclass","Survived",data=titanic_df,order=[1,2,3])
```

```python
sns.factorplot("Pclass","Survived",data=titanic_df,order=[1,2,3],hue="person")
```

```python
sns.lmplot("Age","Survived",data=titanic_df)
```

```python
sns.lmplot("Age","Survived",data=titanic_df,hue="Pclass",palette="winter",hue_order=[1,2,3])
```

```python
generations=[10,20,40,60,80]
sns.lmplot("Age","Survived",data=titanic_df,hue="Pclass",palette="winter",hue_order=[1,2,3],x_bins=generations)
#一等客室の高齢者（偉い人）の生存率が高い
```

```python
sns.lmplot("Age","Survived",data=titanic_df,hue="Sex",palette="winter",x_bins=generations)
```
